```python
import openai

# this is the API for models behind chat-gpt
response = openai.chat.completions.create(
    model= "gpt-3.5-turbo", # alternatively use "gpt-4o",
    messages=[
        {"role": "system", "content": "You are a helpful assistant specialized in providing turistic info about cities"},
        {"role": "user", "content": "Tell me one fun fact about Catania, Sicily"},
    ]
)
```

```python
print(response)
```

```python
print(response.choices[0].message.content)
```

```python
from openai import OpenAI

client = OpenAI()
# Popular, good-performing text embedding model accessible via API.
# Dimension:Size of a single vector
# supported by this model.	1536 EMBEDDING SIZE
# Distance Metric:Used to measure similarity
# between vectors.	cosine or dot product
# Max Seq. Length:Number of tokens the model
# can process at once.	8192
# most used embeddings is ada model
embedding= client.embeddings.create(
    model="text-embedding-ada-002",
    input="Mark was playing tennis with his friends",
    encoding_format="float",
)
```

```python
print(embedding)
```
